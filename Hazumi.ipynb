```python
#dumpfileでの対話１対話開始時刻,　dumpfileでの対話１終了時刻，　dumpfileでの対話１終了行数，　設定範囲
def Fun(start0,end0,end_item_flag,DIFF):
  #print(start,end,end_item_flag)
  #input()
  start_diff_min = 10000
  end_diff_min = 10000
  start_item = -1
  end_item = -1
                                           #二番目から最後までの発話を同じように処理
  #for i in range(rows2)
  for i in range(end_item_flag,rows2): #前回発話の終了時刻から処理
    systime = int(df2.iloc[i:i+1,0:]["SysTime"])
    start_diff = systime - start0
    end_diff = systime - end0

    #get start point
    if(start_diff<=DIFF and start_diff>=-DIFF):
      #print(systime,start_diff,i)
      flag_start = systime
      #global start_diff_min
      if(start_diff_min>abs(start_diff)):
        start_diff_min = abs(start_diff)
        start_item = i
    #if(start_item==21415):
      #print("21415has problem,end_diff is ")
      #input(end_diff)
    #get end point
    if(end_diff<=DIFF and end_diff>=-DIFF):
      #print(systime,end_diff,i)
      flag_end = systime
      #global end_diff_min
      if(end_diff_min>abs(end_diff)):
        end_diff_min = abs(end_diff)
        end_item = i
      if(abs(end_diff)>end_diff_min):
        #print("exit",end_diff,end_diff_min," enditem:",end_item)
        #input()
        break
  #start_item = start_item + 2
  #end_item = end_item + 2
  print("start_item,end_item",start_item,end_item) #開始時刻行数、終了時刻行数

  ##if(end_item==-1):
   # end_item,data_section=Fun(start0,end0,end_item_flag,DIFF+25)
  
  end_item_flag = end_item #終了時刻行数を記録
  data_section = df2.iloc[start_item:end_item+1,0:] #行間データをもらう
  return end_item_flag,data_section
```

# 画像化保存

```python
import csv
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
from scipy import misc
import cv2
import os
from sklearn import preprocessing
```

```python
for j in range(77): #発話数
   i = j + 1
   #ユーザの発話時の座標データ（CSV）をDATAFRAMEに格納
   df = pd.read_csv("/content/section"+str(i)+".csv") 
   #本実験で使用しない特徴量（列）を削除
   df = df.drop(df.columns[0:4],axis=1)
   df = df.drop(['Unnamed: 78'],axis=1)
   #フレーム間差分
   n = np.diff(df,n=1,axis=0)
   #絶対値
   n = abs(n)
   #正規化（０−１）
   min_max_scaler = preprocessing.MinMaxScaler()
   n_minmax = min_max_scaler.fit_transform(n)# axis = 1 
   #正規化確認
   print(n_minmax)
   print(np.max(n_minmax, axis=1))
   #画像化のための（0−255）への変換
   img_array = n_minmax * 255
   #data = pd.DataFrame(img_array)
   #変換した画素値を用いて、PIL imageとして保存
   picture = Image.fromarray(img_array)
   picture = picture.convert("L")
   #画像サイズ
   picture = picture.resize((300,300),Image.ANTIALIAS)                                                 #改
   picture.save("/content/drive/MyDrive/Colab Notebooks/CNN_visualization/Hazumi1911データ/30代女性2/kinect/label_1/発話"+str(i)+".png")
```
